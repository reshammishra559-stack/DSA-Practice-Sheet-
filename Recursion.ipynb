```python
1# WAP to print sum of N numbers .

def sum(n):
    if(n==0):
        return 0
    else:
        return n+sum(n-1)

n=int(input("enter a number"))
print("the sum of n numbers ",sum(n))      
        
```

```python
2#WAP to print factorial for N numbers .

def fact(n):
    if(n==0 or n==1):
        return 1 
    else:
        return n*fact(n-1)
n=int(input("Enter a number"))
print("The factorial of n numbers",fact(n))      
```

```python
3# WAP to reverse a string.

def reverse(s):
     if len(s)<=1:
         return s
     else:
         return s[-1]+reverse(s[:-1])

s=input("Enter a string")
print("The reverse is =",reverse(s))
```

```python
4#WAP to reverse a Array or list.

def reverse_list(list):
    if len(list) <=1:
        return list
    else:
        return [list[-1]]+reverse_list(list[:-1])

l=[1,2,3,4,5]
print("The reverse of the list",reverse_list(l))
```

```python
5# WAP to reverse a numbers.

def reverse(n):
     s=str(n)
     if len(s)<=1:
         return s
     else:
         return s[-1]+reverse(s[:-1])

n=input("Enter a number")
print("The reverse is =",reverse(n))
```

```python
6#WAP to fibonacci number .

def fibo(n):
    if (n==0):
        return 0
    elif(n==1):
        return 1
    else:
        return fibo(n-1)+fibo(n-2)

n=int(input("Enter a Fibo numbers"))
print("the fibonacci number is =",fibo(n))
```
